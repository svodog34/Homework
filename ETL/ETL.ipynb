```python
Player_list = []
Pos_list = []
PTS_list = []
TRB_list = []
AST_list = []
STL_list = []
BLK_list = []


fd_mean = new_NBA_data_2017_df["PTS"].mean()

for index, row in new_NBA_data_2017_df.iterrows():
    if pd.isnull(row["TRB"]):
        Player_list.append(row["Player"])
        Pos_list.append(row["Pos"])
        PTS_list.append(row["PTS"])
        TRB_list.append(row["TRB"])
        AST_list.append(row["AST"])
        STL_list.append(row["STL"])
        BLK_list.append(row["BLK"])
        
        if pd.isnull(row["PTS"]):
            _list.append(fd_mean)
        else:
            PTS_list.append(row["PTS"])
    else:
        Player_list.append(row["Player"])
        Pos_list.append(row["Pos"])
        PTS_list.append(row["PTS"])
        TRB_list.append(row["TRB"])
        AST_list.append(row["AST"])
        STL_list.append(row["STL"])
        BLK_list.append(row["BLK"])
    if pd.isnull(row["PTS"]):
        PTS_list.append(fd_mean)
    else:
        PTS_list.append(row["PTS"])    
```

```python
new_NBA_data_2017_df = pd.DataFrame({
    "Player": Player_list,
    "Pos": Pos_list,
    "PTS": PTS_list,
    "TRB": TRB_list,
    "AST": AST_list,
    "STL": STL_list,
    "BLK": BLK_list
})

new_NBA_data_2017_df.info()
```
